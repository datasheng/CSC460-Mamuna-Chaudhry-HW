```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("titanic.csv")
```

```python
print(df.info())
```

```python
#histograms for the numerical data attributes in the dataset
numerical_cols = df.select_dtypes(include=["int64", "float64"]).columns.tolist() #selecting all numerical datatypes and adding to a column
#function to visualize each type of numerical data attribute with a histogram
for col in numerical_cols:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[col], bins=20, kde=True)
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
#bar chart for categorical data attributes in the dataset
categorical_cols=df.select_dtypes(include=["object"]).columns.tolist()
for col in categorical_cols:
    plt.figure(figsize=(6, 4))
    sns.countplot(x=df[col],
                  order=df[col].value_counts().index,
                  palette="tab10")
    plt.title(f"Count of {col}")
    plt.xticks(rotation=45)
    plt.show()
```

```python
numerical_cols = df.select_dtypes(include=["int64", "float64"]) #selecting all numerical datatypes and adding to a column
max_values = numerical_cols.max()
min_values = numerical_cols.min()
means = numerical_cols.mean()
medians = numerical_cols.median()

print("Maximum Values:\n", max_values)
print("\nMinimum Values:\n", min_values)
print("\nMean Values:\n", means)
print("\nMedian Values:\n", medians)
```

```python
# IQR method for finding outliers in the numerical data attributes
Q1 = df[numerical_cols].quantile(0.25)
Q3 = df[numerical_cols].quantile(0.75)
IQR = Q3 - Q1
# Identify outliers
outliers = ((df[numerical_cols] < (Q1 - 1.5 * IQR)) | (df[numerical_cols] > (Q3 + 1.5 * IQR)))
outlier_dict = {}

for column in numerical_cols:
    outliers_in_column = outliers[column]
    outlier_dict[column] = df[outliers_in_column]
    print(f"\nOutliers in {column}:")
    print(df[outliers_in_column])

print(outlier_dict)

# outliers.shape
```

```python
#boxplot of all numerical variables
numerical_cols = df.select_dtypes(include=["int64", "float64"]).columns.tolist() #selecting all numerical datatypes and adding to a column
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[numerical_cols])
plt.xticks(rotation=90)
plt.title("Boxplot of Statistical Titanic Data")
plt.show()
```

```python
#pairwise scatter plot matrix for numerical data attributes
sns.pairplot(df[numerical_cols])
plt.show()
```

```python
#heatmap for aggregated counts for categorical vs. categorical variables
#categorical data attributes: name,sex,ticket,cabin,embarked

crosstab = pd.crosstab(df["Survived"], df["Pclass"])
plt.figure(figsize=(8, 6))
sns.heatmap(crosstab, annot=True, cmap="Blues", linewidths=0.5)
plt.title("Heatmap of Survived vs Pclass")
plt.show()
```

```python
#visualizing categorical vs numerical data attribute
plt.figure(figsize=(8, 5))
sns.boxplot(x=df["Pclass"], 
            y=df["Fare"],
            palette="Set2")
plt.title("Fare by Pclass")
plt.xticks(rotation=45)
plt.show()
```

```python
#boxplot for categorical and numerical data atributes
#im using Gender and Age
plt.figure(figsize=(8, 5))
sns.barplot(x=df["Sex"], 
            y=df["Age"],
            palette="Set2")
plt.title("Age by Gender")
plt.xticks(rotation=45)
plt.show()
```
